# Data Preprocessing

## Tasks Performed
- Handling missing values
- Converting datetime columns
- Feature engineering
- Data transformation
- Preparing dataset for analysis

```python
# ==========================================
# Import Libraries
# ==========================================

import pandas as pd
import numpy as np

# Ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# ==========================================
# Load Dataset
# ==========================================

df = pd.read_csv(r"C:\Users\hp\Downloads\Real Time Air Quality Index.csv")

print("Dataset Loaded Successfully")
```

```python
# ==========================================
# Display Dataset
# ==========================================

df.head()
```

```python
# ==========================================
# Missing Values
# ==========================================

df.isnull().sum()
```

```python
# ==========================================
# Fill Missing Values
# ==========================================

df['pollutant_min'].fillna(df['pollutant_min'].mean(), inplace=True)

df['pollutant_max'].fillna(df['pollutant_max'].mean(), inplace=True)

df['pollutant_avg'].fillna(df['pollutant_avg'].mean(), inplace=True)
```

```python
# ==========================================
# Verify Missing Values
# ==========================================

df.isnull().sum()
```

```python
# ==========================================
# Convert Date Column
# ==========================================

df['last_update'] = pd.to_datetime(
    df['last_update'],
    format='%d-%m-%Y %H:%M:%S'
)

print(df['last_update'].dtype)
```

```python
# ==========================================
# Feature Engineering
# ==========================================

df['year'] = df['last_update'].dt.year
df['month'] = df['last_update'].dt.month
df['day'] = df['last_update'].dt.day
df['hour'] = df['last_update'].dt.hour

df.head()
```

```python
# ==========================================
# Data Types
# ==========================================

df.dtypes
```

```python
# ==========================================
# Save Cleaned Dataset
# ==========================================

df.to_csv("cleaned_air_quality.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```
