```python
# ==========================================
# Load Saved Model
# ==========================================

import joblib
import pandas as pd

# Load Model
model = joblib.load("../models/air_quality_model.pkl")

# Load Scaler
scaler = joblib.load("../models/scaler.pkl")

print("Model and Scaler Loaded Successfully")
```

```python
# ==========================================
# Test Prediction
# ==========================================

sample_data = pd.DataFrame([[
    45,   # pollutant_min
    120,  # pollutant_max
    11,   # month
    15,   # day
    10    # hour
]], columns=[
    'pollutant_min',
    'pollutant_max',
    'month',
    'day',
    'hour'
])

sample_scaled = scaler.transform(sample_data)

prediction = model.predict(sample_scaled)

print("Predicted Pollution Average:", prediction[0])
```
